```python
import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Review Ratings'] = df.groupby('caegory')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace('','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age_group
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['Age'], q=4, labels = labels)
```

```python
df[['Age','age_group']].head(10)
```

```python
# create column purchase_frequency_days

Frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quaterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['Purchase Frequency Days'] = df['Frequency of Purchases'].map(Frequency_mapping)
```

```python
df[['Purchase Frequency Days','Frequency of Purchases']].head(10)
```
